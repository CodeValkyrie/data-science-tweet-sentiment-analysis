```python
# I think we may need database software to help the data storage and extract which is efficient.
# please download the mongodb from the offical website: https://www.mongodb.com/
import json
import pymongo # make sure install these library
from pprint import pprint
```

```python

### import the json file to the mongoDB ###
# Making a Connection with MongoClient
collection = None
try:
    client = pymongo.MongoClient('mongodb://localhost:27017/')
    # get a database
    db = client['assign1']
    # create and the the collection, 
    collection = db['tweet_test'] # change it to the real one

except Exception as e:
    print(e)
    
# MongoDB strcuture 
# database -> collection -> (key,value)
```

```python
# Read the file line by line and insert to the collection
f = open('test.json', 'r') # this is top 12 tweets for test
# f = open('xaa', 'r') # this is 1/10 of the real dataset 
dataset = f.readlines()
for item in dataset:
    obj = json.loads(item)
    data = collection.insert_one(obj)
```

```python
# View one of the example
top1_tweets = collection.find({}).limit(1)
No_X_tweets = collection.find({}).skip(5).limit(1) # the No.5 tweets
pprint(No_X_tweets[0])
```

```python
# the number of tweets
num_tweets = collection.find({}).count()
num_tweets
```

```python
# return some of the attributes eg: id and text
id_text_tweets = collection.find({},{'_id':0,'id':1,'text':1})
for item in id_text_tweets:
    print(item)
```
